```python
# 데이터를 다운로드 해보자

import os
import tarfile
import urllib.request   #인터넷 창을 여는 화장자

# 다운 받을 주소 루트
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/"
# 데이터 저장할 위치 
HOUSING_PATH = os.path.join("dataset", "housing")
# 저장할 파일 설정
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):  # 현재위치에 파일이잉슨지 보기위한 확인
        os.makedirs(housing_path) #폴더 생성
    tgz_path = os.path.join(housing_path, "housing.tgz") # 자장 위치
    # URL로 표시된 네트워크 객체를 로컬 파일로 복사합니다
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path) #앞축 해제
    housing_tgz.close() #마무리
```

```python
fetch_housing_data()
# 함수 실행
```

```python
import pandas as pd
# 판다스를 이용하여 데이터를 읽어 보겠다
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv") #경로 성정
    return pd.read_csv(csv_path)  #판다스로 데이터 읽기
```

```python
housing = load_housing_data()  #실행, 읽어 드리는 함수를 사용해서 그 자체를 housing 변수에게 줌
housing.head()   # 데이터들의 대가리 읽기
```
