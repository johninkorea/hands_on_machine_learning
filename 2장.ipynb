```python
# 데이터를 다운로드 해보자

import os
import tarfile
import urllib.request   #인터넷 창을 여는 화장자

# 다운 받을 주소 루트
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/"
# 데이터 저장할 위치 
HOUSING_PATH = os.path.join("dataset", "housing")
# 저장할 파일 설정
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):  # 현재위치에 파일이잉슨지 보기위한 확인
        os.makedirs(housing_path) #폴더 생성
    tgz_path = os.path.join(housing_path, "housing.tgz") # 자장 위치
    # URL로 표시된 네트워크 객체를 로컬 파일로 복사합니다
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path) #앞축 해제
    housing_tgz.close() #마무리
```

```python
fetch_housing_data()
# 함수 실행
```

```python
# 데이터의 테스트세트를 나누자
import numpy as np

# 다음의 과정은 numpy의 랜덤을 사용하는데 매번 다른 결과가 나오면 결과를 보기 힘드므로 seed값을 정해주겠다.
np.random.seed(42)



def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) #1
    # permutation: 셔플은 그 자체를 바꾸는데 이함수는 그 자체를 그대로 두고 그 값을 변수에 저장함다
    test_set_size = int(len(data) * test_ratio) #2
    test_indices = shuffled_indices[:test_set_size]  #3
    train_indices = shuffled_indices[test_set_size:]  #3
    return data.iloc[train_indices], data.iloc[test_indices]  #4

# 현재 이함수의 과정을 전체적으로 보면 
# 먼저 전체를 permutation을 하고 원하는 만큼 파일의 크기를 설정한다 1,2 번
# 각 훈련 테스트 셋을 설정 한 크기별로 부류한다. 3번
# 마지막으로 해당 리스트에 맞게 데이터 프래임을 행별로 리턴한다.

# 그니까 대가리만 가져왔다가 나중에 그에 맞는 성분가지 모두 가져오는
```

```python
train_set, test_set = split_train_test(housing, 0.2)
a=len(test_set)
b=len(train_set)
c=(a+b)

print(a,b,c)
```

```python
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```
